```python
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup as bs
from urllib.request import urlretrieve
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import re
import os
import requests as req
import time
import pandas as pd
```

```python
def crawling_restaurant(idx):
    rest_df = pd.DataFrame(index=range(0, 38867), columns=['title', 'address', 'operation', 'contact', 'location_detail', 'hashtag', 'menu'])

    global driver
    time.sleep(2)

    try:
        element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '///*[@id="mArticle"]/div[1]/div[1]/div[2]/div/h2')))
    except:
        pass

    title = driver.find_element(By.XPATH, '//*[@id="mArticle"]/div[1]/div[1]/div[2]/div/h2').text
    rest_df.loc[idx, 'title'] = title

    try:
        address = driver.find_element(By.XPATH, '//*[@id="mArticle"]/div[1]/div[2]/div[1]/div/span[1]').text
        rest_df.loc[idx, 'address'] = address
    except:
        pass

    # 운영시간
    try:
        operation_button = driver.find_element(By.XPATH, '//*[@id="mArticle"]/div[1]/div[2]/div[2]/div/div[1]/ul/li/a')
        operation_button.click()
        time.sleep(1)
    except:
        pass

    #operation = driver.find_element(By.XPATH, '//*[@id="mArticle"]/div[1]/div[2]/div[2]/div/div[2]/div/ul[1]/li').text
    try:
        operation = driver.find_element(By.XPATH, '//*[@id="mArticle"]/div[1]/div[2]/div[2]/div/div[1]/ul').text
        rest_df.loc[idx, 'operation'] = operation
    except:
        pass

    try:
        operation_close = driver.find_element(By.XPATH, '//*[@id="mArticle"]/div[1]/div[2]/div[2]/div/div[2]/div/a')
        operation_close.click()
        time.sleep(0.5)
    except:
        pass

    # 전화번호
    try:
        contact = driver.find_element(By.XPATH, '//*[@id="mArticle"]/div[1]/div[2]/div[3]/div/div/span/span[1]').text
        rest_df.loc[idx, 'contact'] = contact
    except:
        pass

    # 장소 세부사항
    try:
        location_detail = driver.find_element(By.XPATH, '//*[@id="mArticle"]/div[1]/div[2]/div[4]/div').text
        rest_df.loc[idx, 'location_detail'] = location_detail
    except:
        pass

    # 해시태그
    try:
        tag = driver.find_elements(By.XPATH, '//*[@id="mArticle"]/div[1]/div[2]/div[4]/div/div/span/a')
        tag_lst = [i.text for i in tag]
        rest_df.at[idx, 'hashtag'] = tag_lst
    except:
        pass

    # 메뉴
    try:
        menu_more_button = driver.find_element(By.XPATH, '//*[@id="mArticle"]/div[3]/a/span[1]')
        menu_more_button.click()
    except:
        pass

    time.sleep(1)

    try:
        menu = driver.find_elements(By.XPATH, '//*[@id="mArticle"]/div[3]/ul/li/div')
        menu_lst = [i.text for i in menu]
        rest_df.at[idx, 'menu'] = menu_lst
    except:
        pass

    # 사진
    path = 'C:/Users/user21/MiniProject/crawling/image'
    #path = 'D:/ubion/MiniProject/crawling/image'
    try:
        element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//*[@id="mArticle"]/div[5]/div[2]/ul/li/a')))
        ac = ActionChains(driver)
        ac.move_to_element(element)
        ac.perform()
        img_list = driver.find_elements(By.XPATH, '//*[@id="mArticle"]/div[5]/div[2]/ul/li/a')
        img_src = [re.sub('url\(\"|\[\'|\"\)', "", i.value_of_css_property("background-image")) for i in img_list]

        if not os.path.isdir(f'{path}/{title}'):
            os.mkdir(f'{path}/{title}')

        for i in range(len(img_src)):
            urlretrieve(img_src[i], f'{path}/{title}/{title+str(i+1)}.jpg')

        
    except:
        try:
            element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//*[@id="mArticle"]/div[4]/div[2]/ul/li/a')))
            ac = ActionChains(driver)
            ac.move_to_element(element)
            ac.perform()
            img_list = driver.find_elements(By.XPATH, '//*[@id="mArticle"]/div[4]/div[2]/ul/li/a')
            img_src = [re.sub('url\(\"|\[\'|\"\)', "", i.value_of_css_property("background-image")) for i in img_list]

            if not os.path.isdir(f'{path}/{title}'):
                os.mkdir(f'{path}/{title}')

            for i in range(len(img_src)):
                urlretrieve(img_src[i], f'{path}/{title}/{title+str(i+1)}.jpg')
            
        except:
            try:
                element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//*[@id="mArticle"]/div[6]/div[2]/ul/li/a')))
                ac = ActionChains(driver)
                ac.move_to_element(element)
                ac.perform()
                img_list = driver.find_elements(By.XPATH, '//*[@id="mArticle"]/div[4]/div[2]/ul/li/a')
                img_src = [re.sub('url\(\"|\[\'|\"\)', "", i.value_of_css_property("background-image")) for i in img_list]

                if not os.path.isdir(f'{path}/{title}'):
                    os.mkdir(f'{path}/{title}')

                for i in range(len(img_src)):
                    urlretrieve(img_src[i], f'{path}/{title}/{title+str(i+1)}.jpg')
            except:
                pass

```

```python
keyword = '이수역 학원'
url = 'https://map.kakao.com/'
driver = wb.Chrome()
driver.get(url)
driver.maximize_window()
idx = 0


input_box = driver.find_element(By.XPATH, '//*[@id="search.keyword.query"]')
input_box.send_keys(keyword)
input_box.send_keys(Keys.ENTER)

try:
    click_blue_popup = driver.find_element(By.CSS_SELECTOR, 'body > div.coach_layer.coach_layer_type1 > div > div > div > span')
    click_blue_popup.click()
except:
    pass

try:
    loc_more = driver.find_element(By.XPATH, '//*[@id="info.search.place.more"]')
    time.sleep(1)
    loc_more.click()
except:
    loc_more = driver.find_element(By.XPATH, '//*[@id="info.search.place.more"]')
    time.sleep(1)
    loc_more.click()

page_buttons = driver.find_elements(By.CSS_SELECTOR, '#info\.search\.page > div > a')
time.sleep(1)
page_buttons[0].click()

while True:
    for i in range(1, 6):
        if (i < 5):
            driver.switch_to.window(driver.window_handles[0])
            time.sleep(1)
            page_buttons[i].click()
            
            time.sleep(1)
            more_view_list = driver.find_elements(By.CLASS_NAME, 'moreview')

            for i in range(len(more_view_list)):
                driver.switch_to.window(driver.window_handles[0])
                time.sleep(1)
                more_view_list[i].click()

                driver.switch_to.window(driver.window_handles[-1])
                crawling_edu(idx)
                driver.close()
                idx += 1
        
        else:
            driver.switch_to.window(driver.window_handles[0])
            click_next = driver.find_element(By.XPATH, '//*[@id="info.search.page.next"]')
            click_next.click()
            time.sleep(1)
```
