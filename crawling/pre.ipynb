```python
import pandas as pd
import numpy as np
import random
from datetime import datetime
```

```python
food_df = pd.read_csv('origin_data/food_df.csv', encoding='cp949')
now = datetime.now()
```

```python
food_df['hashtag'] = food_df['hashtag'].apply(lambda x : x if len(x) > 2 else np.nan)
food_df['menu'] = food_df['menu'].apply(lambda x : x if len(x) > 2 else np.nan)
food_df.drop('location_detail', axis=1, inplace=True)
food_df.drop('hashtag', axis=1, inplace=True)
food_df.reset_index(inplace=True)

mem_seq = [random.randrange(1, 21) for i in range(len(food_df))]
food_df.insert(1, 'mem_seq', mem_seq)
res_score = [random.randrange(1, 6) for i in range(len(food_df))]
food_df.insert(len(food_df.columns), 'res_score', res_score)
food_df.insert(len(food_df.columns), 'res_hit', np.zeros(len(food_df)))
food_df.insert(len(food_df.columns), 'res_wdate', now.strftime('%Y-%m-%d %H:%M:%S'))
food_df.insert(len(food_df.columns), 'res_update', now.strftime('%Y-%m-%d %H:%M:%S'))
food_df['res_hit'].astype(int)
food_df.columns = ['res_seq', 'mem_seq', 'res_name', 'res_locate', 'res_content', 'res_phone', 'menu', 'res_score', 'res_hit', 'res_wdate', 'res_update']
food_df['res_content'] = food_df['res_content'].fillna('hi')
food_menu_df = food_df[['res_seq', 'mem_seq', 'menu']]

food_df.drop('menu', axis=1, inplace=True)
food_df.columns
```

```python
res_phone_fillna = []
for i in range(len(food_df[food_df['res_phone'].isnull()])):
    temp_phone = [str(random.randrange(0, 10)) for i in range(8)]
    res_phone_fillna.append('02-'+''.join(temp_phone[:4])+'-'+''.join(temp_phone[4:]))
for i in range(len(food_df)):
    if pd.isnull(food_df.loc[i, 'res_phone']):
        food_df.loc[i, 'res_phone'] = res_phone_fillna.pop()
```

```python
food_df = food_df[['res_seq', 'mem_seq', 'res_name', 'res_locate', 'res_phone','res_content',
                   'res_score', 'res_hit', 'res_wdate', 'res_update']]
```

```python
food_df['res_content'] = food_df['res_content'].apply(lambda x : x.replace('\n닫기', ''))
```

```python
food_df['res_hit'] = food_df['res_hit'].apply(lambda x: int(x))
food_df
```

```python
food_df.to_csv('pre_data\\food_df.csv', encoding='cp949', index=False)
```

```python
food_df = pd.read_csv('origin_data/food_df.csv', encoding='cp949')

food_df['hashtag'] = food_df['hashtag'].apply(lambda x : x if len(x) > 2 else np.nan)
food_df['menu'] = food_df['menu'].apply(lambda x : x if len(x) > 2 else np.nan)
food_df.drop(['location_detail', 'operation', 'contact', 'address', 'title'], axis=1, inplace=True)
food_df.drop('hashtag', axis=1, inplace=True)
food_df.reset_index(inplace=True)

food_df.columns = ['res_seq', 'menu']
food_menu_df = food_df
food_pre_df = pd.read_csv('pre_data\\food_df.csv', encoding='cp949')
```

```python
food_menu_df = pd.merge(food_pre_df, food_menu_df, how='inner', on='res_seq')
food_menu_df = food_menu_df[['res_seq', 'mem_seq', 'menu']]
```

```python
food_menu_df = food_menu_df[food_menu_df['menu'].notnull()]
```

```python
food_menu_df.reset_index(drop=True, inplace=True)
```

```python
for i in range(len(food_menu_df)):
    food_menu_df.loc[i,'menu'] = food_menu_df.loc[i,'menu'].split('\,\'')[0][1:-1]
food_menu_df
```

```python
food_menu_df_temp = food_menu_df.loc[:,'res_seq':'mem_seq']
food_menu_df_temp.insert(2, "menu", "") 
food_menu_df_temp.insert(2, "res_item_title", "") 
food_menu_df_temp.insert(2, "res_item_price", "") 
```

```python
food_menu_df_temp
```

```python
food_menu_df_final = pd.DataFrame(index=range(0, 5000), columns=['res_seq', 'mem_seq', 'menu'])
idx = 0
for i in range(len(food_menu_df)):
    
    for j in food_menu_df.loc[i,'menu'].split("',"):
        food_menu_df_final.loc[idx, 'res_seq'] = food_menu_df_temp.loc[i,'res_seq']
        food_menu_df_final.loc[idx, 'mem_seq'] = food_menu_df_temp.loc[i,'mem_seq']
        food_menu_df_final.loc[idx, 'menu'] = j
        idx += 1

food_menu_df_final.dropna(inplace=True)
food_menu_df_final
```

```python
for i in range(len(food_menu_df_final)):
    if len(food_menu_df_final.loc[i, 'menu'].split('\\n')) >= 2:
        food_menu_df_final.loc[i, 'res_item_title'] = food_menu_df_final.loc[i, 'menu'].split('\\n')[0]
        food_menu_df_final.loc[i, 'res_item_price'] = food_menu_df_final.loc[i, 'menu'].split('\\n')[1]
food_menu_df_final
```

```python
import re
food_menu_df_final.dropna(inplace=True)
food_menu_df_final['res_item_price'] = food_menu_df_final['res_item_price'].apply(lambda x: re.sub(r'[^0-9]', '', x))
food_menu_df_final['res_item_price'] = food_menu_df_final['res_item_price'].apply(lambda x: np.nan if len(x)< 1 else x)
food_menu_df_final.dropna(inplace=True)
```

```python
food_menu_df_final.reset_index(inplace=True)
food_menu_df_final.drop('index', axis=1, inplace=True)
food_menu_df_final.reset_index(inplace=True)
```

```python
food_menu_df_final
```

```python
food_menu_df_final.drop('menu', axis=1, inplace=True)
food_menu_df_final.columns = ['res_item_seq', 'res_seq', 'mem_seq', 'res_item_title', 'res_item_price']
food_menu_df_final = food_menu_df_final[['res_item_seq', 'res_seq', 'mem_seq', 'res_item_title', 'res_item_price']]
food_menu_df_final.insert(4, 'res_item_content', '맛있따')
food_menu_df_final.insert(4, 'res_item_pic', '')
food_menu_df_final.insert(len(food_menu_df_final.columns), 'res_item_wdate', now.strftime('%Y-%m-%d %H:%M:%S'))
food_menu_df_final.insert(len(food_menu_df_final.columns), 'res_item_update', now.strftime('%Y-%m-%d %H:%M:%S'))
```

```python
food_menu_df_final['res_item_title'] = food_menu_df_final['res_item_title'].apply(lambda x: x.replace("'", '').lstrip())
food_menu_df_final['res_item_price'] = food_menu_df_final['res_item_price'].apply(lambda x: int(x))
food_menu_df_final = food_menu_df_final[food_menu_df_final['res_item_price'] > 500]
```

```python
food_menu_df_final['res_seq'] = food_menu_df_final['res_seq'].apply(lambda x: int(x))
food_menu_df_final['mem_seq'] = food_menu_df_final['mem_seq'].apply(lambda x: int(x))
food_menu_df_final = food_menu_df_final[['res_item_seq', 'res_seq', 'mem_seq', 'res_item_title', 'res_item_content', 'res_item_pic', 'res_item_price', 'res_item_wdate', 'res_item_update']]
```

```python
food_menu_df_final.to_csv('pre_data\\food_menu_df_temp.csv', encoding='cp949', index=False)
```
